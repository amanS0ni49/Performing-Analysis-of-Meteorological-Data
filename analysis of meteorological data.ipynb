# 1. Import all the required libraries for analysis.

Pandas: Specially used for data analysis and allows us to import/export, data manipulation, and data wrangling.
Numpy: Used for faster mathematical operation on arrays.
Matplotlib: Used for beautiful graphs and data visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 2. Data reading, cleaning, and exploration.

```python
df=pd.read_csv("weatherHistory.csv")
```

```python
df.head()
```

```python

titles_req = ["Formatted Date", "Apparent Temperature (C)", "Humidity"]
df = df[titles_req]
```

```python

df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df = df.set_index('Formatted Date')
df = df.resample('MS').mean()
```

```python
df.head()
```

# 3. Plotting the average temperature and average humidity

```python
plt.figure(figsize=(14,6))
plt.title("Variation in Apparent Tempearture and Humidity with time")
plt.plot(df)
```

```python

df_april = df[df.index.month==4]
```

```python

plt.figure(figsize=(14,6))
plt.plot(df_april)
```
